```python
import matplotlib.pyplot as plt

import torch
from torchvision import datasets, transforms
from torchvision.io import read_image
```

```python
import os
import numpy as np
import shutil

path='/content/Combined/colored dataset'
test_path='/content/Combined/test'

if not os.path.exists(test_path):
  os.mkdir(test_path)
				
folder_enum = ['01','02','03','04','05','06','07','08', '09', '10']

for label_id in folder_enum:
  images = np.array(os.listdir(path + '/' + label_id))
  np.random.permutation(images)
  if not os.path.exists(test_path + '/' + label_id):
    os.mkdir(test_path + '/' + label_id)
  for i in range(45):
    shutil.move(path + '/' + label_id + '/' + images[i], test_path + '/' + label_id + '/' + images[i])

print(os.listdir(test_path + '/' + label_id))
```
