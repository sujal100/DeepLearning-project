```python
import os
import cv2
import dlib 
import keras
import imutils
import numpy as np 
import pandas as pd 
from imutils import face_utils

def rect_to_bb(rect):
    
	#	boundary prediction with the help of dlib into a 4 tuple format (a,b,c,d)
	left = rect.left()
	top = rect.top() 
	right = rect.right() - left
	bottom = rect.bottom() - top

	return (left, top, right, bottom)

def shape_to_tuple(shape, dtype="int"):
	
	coords = np.zeros((68, 2), dtype=dtype)

	
	#	conversion of the 68 facial landmarks to 2 tuple coordinates
	for i in range(0, 68):
		coords[i] = (shape.part(i).x, shape.part(i).y)

	# return the list of (x, y)-coordinates
	return coords

def crop_and_save_image(img, img_path, write_img_path, img_name):
	detector = dlib.get_frontal_face_detector()

	predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat') # We found the file with the dataset
	

	#	loading the image with the help of cv2
	image = cv2.imread(img_path)
	gray=image

	#	detect faces in the grayscale image; detecting the validity of 1 face
	rects = detector(gray, 1)
	if len(rects) > 1:
		print( "ERROR: more than one face detected")
		return
	if len(rects) < 1:
		print( "ERROR: no faces detected")
		return


	for (i, rect) in enumerate(rects):
		shape = predictor(gray, rect)
		shape = face_utils.shape_to_tuple(shape)
		name, i, j = 'mouth', 48, 68

		(x, y, w, h) = cv2.boundingRect(np.array([shape[i:j]]))
		print(x,y,w,h)		#printing the rect boundings

		padding_h=50-h
		padding_w=50-w

		if padding_h%2 ==1:
			padding_h_up=int(padding_h/2)
			padding_h_down=padding_h_up+1
		else:
			padding_h_up=int(padding_h/2)
			padding_h_down=padding_h_up

		y=y-padding_h_up
		h=h+padding_h_down

		if padding_w%2 ==1:
			padding_w_left=int(padding_w/2)
			padding_w_right=padding_w_left+1
		else:
			padding_w_left=int(padding_w/2)
			padding_w_right=padding_w_left

		x=x-padding_w_left
		w=w+padding_w_right
		w=h=50

		#boundings after resizing 
		print(x,y,w,h)
		roi = gray[y:y+h, x:x+w]
		roi = cv2.resize(roi, (100,100),interpolation = cv2.INTER_AREA) 
		       
		print('C:/Users/Praneeth/Deep Learning/Project/colored cropped/' + write_img_path)
		#print(np.array(roi))


		if not cv2.imwrite('C:/Users/Praneeth/Deep Learning/Project/colored cropped/' + write_img_path, roi):
			raise Exception("Couldn't write")
		break	

os.listdir('D:/3.2/DL/Project/dataset')
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')  # We found the file with the dataset

people = ['M01','M02']
data_types = ['words']

folder_enum = ['01','02','03','04','05','06','07','08', '09', '10']
instances = ['01','02','03','04','05','06','07','08', '09', '10']


words = ['Begin', 'Choose', 'Connection', 'Navigation', 'Next', 'Previous', 'Start', 'Stop', 'Hello', 'Web'] 

words_di = {i:words[i] for i in range(len(words))}

if not os.path.exists('colored cropped'):
    os.mkdir('colored cropped')


for person_ID in people:
	if not os.path.exists('colored cropped/' + person_ID ):
		os.mkdir('colored cropped/' + person_ID)

	for data_type in data_types:
		if not os.path.exists('colored cropped/' + person_ID + '/' + data_type):
			os.mkdir('colored cropped/' + person_ID + '/' + data_type)

		for phrase_ID in folder_enum:
			if not os.path.exists('colored cropped/' + person_ID + '/' + data_type + '/' + phrase_ID):
				os.mkdir('colored cropped/' + person_ID + '/' + data_type + '/' + phrase_ID)

			for instance_ID in instances:
                
				if not os.path.exists('colored cropped/' + person_ID + '/' + data_type + '/' + phrase_ID + '/' + instance_ID):
					os.mkdir('colored cropped/' + person_ID + '/' + data_type + '/' + phrase_ID + '/' + instance_ID)

				directory = 'D:/3.2/DL/Project/dataset/' + person_ID + '/' + data_type + '/' + phrase_ID + '/' + instance_ID + '/'
				dir_temp = person_ID + '/' + data_type + '/' + phrase_ID + '/' + instance_ID + '/'
				filelist = os.listdir(directory)
				for img_name in filelist:
					if img_name.startswith('color'):
						image = cv2.imread(directory + '' + img_name)
						crop_and_save_image(image, directory + '' + img_name,dir_temp + '' + img_name, img_name)
```
